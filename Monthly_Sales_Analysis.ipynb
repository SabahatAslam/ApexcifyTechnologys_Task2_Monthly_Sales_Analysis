### 🔑Key Concepts

- Data visualization, plot(), xlabel(), ylabel(), title()

```python
import matplotlib.pyplot as plt
import pandas as pd
monthly_sales = pd.read_csv("/content/Monthly_sales_dataset.csv") #paste the file path
monthly_sales
```

### 📈Data Visualization Libraries

For visualizing data, we mainly use:

- Matplotlib (for basic and simple plots)
- Seaborn (for complex plots with availability of beautiful themes and paletes)

We  will plot a line chart.
>💡Line charts are ideal for trend analysis 📈.

```python
 # Here we use matplotlib
plt.plot(monthly_sales["Month"], monthly_sales["Sales($)"])
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.xticks(rotation=30) # Rotate x-axis labels by 30 degrees

plt.show()
```

### 🔍Monthly Sales Analysis
We can also further dissect our data by month for deep insights.
>💡Bar plot would be best to observe changes in monthly sales 📊.

```python
plt.bar(monthly_sales["Month"], monthly_sales["Sales($)"], color='lightgreen')
plt.title("Monthly Sales Analysis")
plt.ylabel("Sales($)")
plt.xticks(rotation=30)
plt.show()
```

### 📝Sales Analysis Summary:

**Analyzing the sales data:**

The overall trend shows an increase in sales, from a minimum of 12,000 dollars in January to a maximum of 22,000 dollars in December.
Month-by-month analysis shows slight dips in January, March, May, July, September, and November.



The peak sales in December are likely due to the holiday season, while the sharp decline in January may be a result of the post-holiday dip.

**🔑 Key Points**

**1. Lowest Sales**

- January (post-holiday dip)

**2. Highest Sales**

- December (peak holiday season)

**3. Mid-year Trend**

- Steady increase with slight dips (possibly due to seasonal fluctuations or marketing/promotion impact)

**4. Overall Trend**

- Upward trend throughout the year
- Growth is consistent with seasonal peaks at the end of the year
